```python
#tabulation

class Solution:
    # Function to compute the minimum cost to reach the end using at most 'k' jumps
    def solveUtil(self, n, height, dp, k):
        # Base case: cost to reach first stone is 0
        dp[0] = 0

        # Iterate over each stone
        for i in range(1, n):
            # Initialize minimum cost for this stone as large value
            mmSteps = float('inf')

            # Try all possible jump lengths from 1 to k
            for j in range(1, k + 1):
                # Ensure jump doesn't go out of bounds
                if i - j >= 0:
                    # Cost of jumping from (i - j) to i
                    jump = dp[i - j] + abs(height[i] - height[i - j])
                    # Keep track of the minimum cost
                    mmSteps = min(mmSteps, jump)

            # Store computed minimum cost
            dp[i] = mmSteps

        # Last element of dp stores the answer
        return dp[n - 1]

    # Function to solve the problem
    def solve(self, n, height, k):
        # Initialize DP array with -1
        dp = [-1] * n
        return self.solveUtil(n, height, dp, k)

# Driver code
if __name__ == "__main__":
    # Heights of stones
    height = [30, 10, 60, 10, 60, 50]
    n = len(height)
    k = 2

    # Create solution object
    sol = Solution()

    # Output result
    print(sol.solve(n, height, k))
```

```python
#Maximum sum of non-adjacent elements (DP 5) memorixation

class Solution:
    # Recursive helper with memoization
    def solve(self, arr, i, dp):
        # No elements to pick
        if i < 0:
            return 0

        # One element case
        if i == 0:
            return arr[0]

        # Return memoized result
        if dp[i] != -1:
            return dp[i]

        # Pick current and go 2 steps back
        pick = arr[i] + self.solve(arr, i - 2, dp)

        # Skip current and go 1 step back
        notPick = self.solve(arr, i - 1, dp)

        # Store and return max
        dp[i] = max(pick, notPick)
        return dp[i]

    # Main function to call
    def maximumNonAdjacentSum(self, arr):
        n = len(arr)

        # Initialize DP array
        dp = [-1] * n

        # Start recursion
        return self.solve(arr, n - 1, dp)

# Driver code
arr = [2, 1, 4, 9]
obj = Solution()
print(obj.maximumNonAdjacentSum(arr))
```

```python
#tabulzation

class Solution:
    # Function to return maximum sum of non-adjacent elements
    def maximumNonAdjacentSum(self, arr):
        # Get size of array
        n = len(arr)

        # If array has one element
        if n == 1:
            return arr[0]

        # Initialize dp array
        dp = [0] * n

        # Base cases
        dp[0] = arr[0]
        dp[1] = max(arr[0], arr[1])

        # Iterate from 3rd element
        for i in range(2, n):
            # Pick current + dp[i-2] or skip current
            dp[i] = max(arr[i] + dp[i - 2], dp[i - 1])

        # Final result at last index
        return dp[n - 1]


# Driver code
arr = [2, 1, 4, 9]
sol = Solution()
print(sol.maximumNonAdjacentSum(arr))
```

```python
#space optimzatio

class Solution:
    # Function to return maximum sum of non-adjacent elements
    def maxSum(self, nums):
        # Handle edge case for empty input
        if not nums:
            return 0

        # Initialize two tracking sums
        prev2 = 0
        prev = nums[0]

        # Traverse through the array
        for i in range(1, len(nums)):
            # Include current by adding it to two steps back
            include = nums[i] + prev2

            # Exclude current by taking previous best
            exclude = prev

            # Choose max of include and exclude
            curr = max(include, exclude)

            # Update tracking variables
            prev2 = prev
            prev = curr

        # Final result is stored in prev
        return prev


# Driver code
arr = [3, 2, 5, 10, 7]
obj = Solution()
print(obj.maxSum(arr))
```

```python
#house robber

class Solution:
    # Function to solve the linear house robber problem
    def solve(self, arr):
        n = len(arr)

        # If there's only one house, return its value
        if n == 1:
            return arr[0]

        # prev stores the maximum sum till the previous index
        prev = arr[0]

        # prev2 stores the maximum sum till index before previous
        prev2 = 0

        # Iterate from the second house
        for i in range(1, n):
            # Pick the current house
            pick = arr[i]
            if i > 1:
                pick += prev2

            # Skip the current house
            nonPick = prev

            # Choose maximum
            cur_i = max(pick, nonPick)

            # Update prev2 and prev
            prev2 = prev
            prev = cur_i

        return prev

    # Function to solve the circular house robber problem
    def robStreet(self, n, arr):
        # Handle edge cases
        if n == 0:
            return 0
        if n == 1:
            return arr[0]

        # arr1 excludes first house
        arr1 = arr[1:]
        # arr2 excludes last house
        arr2 = arr[:-1]

        return max(self.solve(arr1), self.solve(arr2))


# Driver code
if __name__ == "__main__":
    # Example input
    arr = [1, 5, 1, 2, 6]
    n = len(arr)

    # Create Solution object
    sol = Solution()

    # Output result
    print(sol.robStreet(n, arr))
```

```python
#Subset sum equal to target (DP- 14)

class Solution:
    # Recursive helper function with memoization
    def subsetSumUtil(self, ind, target, arr, dp):
        # Base case: target achieved
        if target == 0:
            return True

        # Base case: first index check
        if ind == 0:
            return arr[0] == target

        # If already computed
        if dp[ind][target] != -1:
            return dp[ind][target] == 1

        # Choice 1: not take the element
        notTaken = self.subsetSumUtil(ind - 1, target, arr, dp)

        # Choice 2: take the element if possible
        taken = False
        if arr[ind] <= target:
            taken = self.subsetSumUtil(ind - 1, target - arr[ind], arr, dp)

        # Store result
        dp[ind][target] = 1 if (notTaken or taken) else 0
        return notTaken or taken

    # Main function to call the helper
    def subsetSumToK(self, n, k, arr):
        dp = [[-1] * (k + 1) for _ in range(n)]
        return self.subsetSumUtil(n - 1, k, arr, dp)


# Driver code
if __name__ == "__main__":
    arr = [1, 2, 3, 4]
    k = 4
    sol = Solution()

    if sol.subsetSumToK(len(arr), k, arr):
        print("Subset with the given target found")
    else:
        print("Subset with the given target not found")
```

```python
class Solution:
    def subsetSumToK(self, n, k, arr):
        # Initialize DP table with False values
        dp = [[False] * (k + 1) for _ in range(n)]

        # Base case: sum 0 can always be formed with empty subset
        for i in range(n):
            dp[i][0] = True

        # Base case: if first element <= k, set that position True
        if arr[0] <= k:
            dp[0][arr[0]] = True

        # Fill the DP table iteratively
        for ind in range(1, n):
            for target in range(1, k + 1):
                # Option 1: not take current element
                not_taken = dp[ind - 1][target]

                # Option 2: take current element if possible
                taken = False
                if arr[ind] <= target:
                    taken = dp[ind - 1][target - arr[ind]]

                # Current cell is True if either option is True
                dp[ind][target] = not_taken or taken

        # Final answer: can sum k be formed using all elements?
        return dp[n - 1][k]


# Driver code
if __name__ == "__main__":
    arr = [1, 2, 3, 4]
    k = 4
    n = len(arr)
    sol = Solution()

    if sol.subsetSumToK(n, k, arr):
        print("Subset with the given target found")
    else:
        print("Subset with the given target not found")
```

```python
#Partition Equal Subset Sum (DP- 15) memo

class Solution:
    # Helper function to check if subset with target sum exists
    def subsetSumUtil(self, ind, target, arr, dp):
        # Base case: target sum reached
        if target == 0:
            return True

        # Base case: first element check
        if ind == 0:
            return arr[0] == target

        # If already computed, return stored result
        if dp[ind][target] != -1:
            return dp[ind][target] == 1

        # Choice 1: exclude current element
        notTaken = self.subsetSumUtil(ind - 1, target, arr, dp)

        # Choice 2: include current element if possible
        taken = False
        if arr[ind] <= target:
            taken = self.subsetSumUtil(ind - 1, target - arr[ind], arr, dp)

        # Store result in dp table
        dp[ind][target] = 1 if (notTaken or taken) else 0
        return notTaken or taken

    # Main function to check if array can be partitioned into two equal subsets
    def canPartition(self, arr):
        # Find array size
        n = len(arr)

        # Calculate total sum
        totSum = sum(arr)

        # If sum is odd, partition not possible
        if totSum % 2 == 1:
            return False

        # Target sum for each subset
        k = totSum // 2

        # DP table initialization
        dp = [[-1] * (k + 1) for _ in range(n)]

        # Check possibility using recursion + memoization
        return self.subsetSumUtil(n - 1, k, arr, dp)


if __name__ == "__main__":
    # Example array
    arr = [2, 3, 3, 3, 4, 5]

    # Create object
    obj = Solution()

    # Output result
    if obj.canPartition(arr):
        print("The Array can be partitioned into two equal subsets")
    else:
        print("The Array cannot be partitioned into two equal subsets")
```

```python

#tabulaton

class Solution:
    # Determines if the given array can be split into two subsets
    # whose sums are equal using a tabulation DP approach
    def canPartition(self, arr):
        n = len(arr)

        # Step 1: Calculate total sum
        totalSum = sum(arr)

        # Step 2: If total sum is odd, partition is impossible
        if totalSum % 2 != 0:
            return False

        # Step 3: Target sum for each subset
        targetSum = totalSum // 2

        # Step 4: Create DP table and initialize
        dp = [[False] * (targetSum + 1) for _ in range(n)]

        # Step 5: Base case: sum 0 is always possible
        for i in range(n):
            dp[i][0] = True

        # Step 6: Initialize first row
        if arr[0] <= targetSum:
            dp[0][arr[0]] = True

        # Step 7: Fill DP table
        for index in range(1, n):
            for target in range(1, targetSum + 1):
                notTaken = dp[index - 1][target]
                taken = False
                if arr[index] <= target:
                    taken = dp[index - 1][target - arr[index]]
                dp[index][target] = notTaken or taken

        # Step 8: Return result
        return dp[n - 1][targetSum]


# Example usage
arr = [2, 3, 3, 3, 4, 5]
solver = Solution()
if solver.canPartition(arr):
    print("The array can be partitioned into two equal subsets")
else:
    print("The array cannot be partitioned into two equal subsets")
```

```python
class Solution:
    # Checks if the given array can be partitioned into two subsets
    # such that the sum of elements in both subsets is equal
    def canPartition(self, arr):
        n = len(arr)

        # Calculate total sum of the array
        totalSum = sum(arr)

        # If the sum is odd, partition into equal subsets is impossible
        if totalSum % 2 != 0:
            return False

        target = totalSum // 2

        # prev[sum] indicates if a subset with 'sum' is possible so far
        prev = [False] * (target + 1)
        prev[0] = True  # sum = 0 is always possible

        # Initialize with the first element
        if arr[0] <= target:
            prev[arr[0]] = True

        # Iterate over all remaining elements
        for i in range(1, n):
            # cur[sum] will store possibilities for current element
            cur = [False] * (target + 1)
            cur[0] = True  # sum 0 always possible

            # Check all possible target sums
            for s in range(1, target + 1):
                # Option 1: Do not take current element
                notTaken = prev[s]

                # Option 2: Take current element if it fits
                taken = False
                if arr[i] <= s:
                    taken = prev[s - arr[i]]

                # Current sum possible if either option is true
                cur[s] = notTaken or taken

            # Move to the next element
            prev = cur

        # Final answer is whether target sum is possible
        return prev[target]


# Example usage
arr = [2, 3, 3, 3, 4, 5]
obj = Solution()
if obj.canPartition(arr):
    print("The Array can be partitioned into two equal subsets")
else:
    print("The Array cannot be partitioned into two equal subsets")
```
